```python
%matplotlib inline

import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv("wineQualityReds.csv")
```

```python
wine.head(n=6)
```

```python
wine=wine.drop('Unnamed: 0', axis=1)#drop a column
```

```python
wine.head()
```

```python
X_wine= wine.drop("quality", axis=1) #predictors
#Y = wine["quality"] #response
```

```python
from sklearn.mixture import GMM      # 1. Choose the model class
model = GMM(n_components=3,
            covariance_type='full')
```

```python
model.fit(X_wine)                    # 3. Fit to data. Notice y is not specified!
```

```python
y_gmm = model.predict(X_wine) 
```

```python
import seaborn as sns
```

```python
wine['cluster'] = y_gmm
```
